```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function to take number of dice side and number of rolls, and return the average roll value and plot a histogram of the rolls
def dice_rolls(num_sides, num_rolls):
    rolls = np.random.randint(1, num_sides + 1, num_rolls)
    avg_roll = np.mean(rolls)
    plt.hist(rolls)
    plt.xlabel('Roll Value')
    plt.ylabel('Number of Rolls')
    plt.title('Dice Rolls')
    plt.show()
    return print(avg_roll)
```

```python
dice_rolls(20, 1000000)
```

10.5 is the expected value of a 20 sided die, and we get a value very close to that. Also, since no sides of the die are favored, we get a uniform distribution of rolls, as expected. 

```python
# function to roll an n-sided die m times and return the maximum value of the rolls

def max_rolls(num_sides, num_rolls):
    rolls = np.random.randint(1, num_sides + 1, num_rolls)
    max_roll = np.max(rolls)
    return max_roll
```

```python
# make a histogram of the max_rolls of 1000000 iterations
max_rolls_list = []
for i in range(1000000):
    max_rolls_list.append(max_rolls(20, 2))
plt.hist(max_rolls_list)
plt.xlabel('Max Roll Value')
plt.ylabel('Number of Rolls')
plt.title('Max Rolls')
plt.show()
```

This histogram makes sense because the function only keeps the larger rolls, so of course you would have more results that are larger.
